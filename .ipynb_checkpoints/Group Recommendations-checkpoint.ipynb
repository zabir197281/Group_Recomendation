```python
"""Create a user-item matrix"""
user_item_matrix = df.pivot_table(index='movie_id', columns='user_id', values='rating')
```

```python
user_item_matrix
```

```python
"""Function to calculate Pearson correlation between users"""
def calculate_pearson_correlation(user_item_matrix):
    user_corr = user_item_matrix.corr(method='pearson', min_periods=5)
    return user_corr
```

```python
correlation_matrix=calculate_pearson_correlation(user_item_matrix).fillna(0)
```

```python
correlation_matrix
```

```python
import numpy as np
user_mean_ratings = user_item_matrix.mean(axis=0)
```

```python
def predict_score(active_user, item, user_item_matrix, correlation_matrix):
    """Calculate mean ratings for each user"""
   
    
    """Get active user mean rating and item ratings by other users"""
    active_user_mean = user_mean_ratings.get(active_user, 0)
    item_ratings = user_item_matrix.loc[item].dropna()
    
    """Get correlations for the active user with others"""
    user_correlations = correlation_matrix.get(active_user, {})
    
    """Sorting the user_correlations matrix"""
    user_correlations=user_correlations.sort_values(ascending=False)
    
    """Sececting the top 50 similar users"""
    user_correlations=user_correlations[0:51,]
    
    """Calculate the weighted sum of deviations from the mea"""
    weighted_sum = sum(user_correlations.get(other_user, 0) * (item_rating - user_mean_ratings.get(other_user, 0))
                       for other_user, item_rating in item_ratings.items()
                       if other_user != active_user)
    
    """Calculate the sum of the absolute values of the correlations"""
    sum_of_correlations = sum(abs(user_correlations.get(other_user, 0))
                              for other_user in item_ratings.index
                              if other_user != active_user)
    
    """Predict the score"""
    if sum_of_correlations == 0:
        """If there are no correlations, return the active user's mean rating"""
        return active_user_mean
    else:
        """Adjust the active user's average by the weighted sum of deviations"""
        prediction = active_user_mean + (weighted_sum / sum_of_correlations)
        
        return  prediction
```

```python
def get_top_similar_users(user_id, correlation_matrix, n=10):
    """Retrieve the correlation series for the user and drop the user's own correlation"""
    user_correlations = correlation_matrix[user_id].drop(user_id)
    """Sort by correlation in descending order and take the top n"""
    top_users = user_correlations.sort_values(ascending=False).head(n)
    return top_users
```

```python
def get_recommended_movies(user_id, user_item_matrix, correlation_matrix):
    """Get the mean ratings for each user"""
    user_mean_ratings = user_item_matrix.mean(axis=1)
    """Get the list of movies the user hasn't rated yet"""
    unrated_movies = user_item_matrix.loc[user_item_matrix[user_id].isna(), :]
    """Dictionary to store the predicted ratings"""
    predictions = {}
    
    for movie_id in unrated_movies.index:
        """Predict the rating for each unrated movie"""
        predicted_rating = predict_score(user_id, movie_id, user_item_matrix, correlation_matrix)
        predictions[movie_id] = predicted_rating
    
    """Sort the predicted ratings in descending order and get the top n movies"""
    recommended_movies = sorted(predictions.items(), key=lambda x: x[1], reverse=True)
    return recommended_movies
```

## Movie score predection for a group of three users with Average Method

```python
User_1 = 196
User_2 = 222
User_3 = 12

group_average_method=[User_1,User_2,User_3]
user_recomendations_average_method={}
```

```python
for user in group_average_method:
    uses_recomendation=get_recommended_movies(user, user_item_matrix, correlation_matrix)
    user_recomendations_average_method[user]=  dict(uses_recomendation)
```

```python
def predected_score_avarage_method(user_recomendations_average_method):
    first_user= user_recomendations_average_method[User_1]
    second_user=user_recomendations_average_method[User_2]
    third_user= user_recomendations_average_method[User_3]
    all_keys=set(first_user.keys()) | set(second_user.keys()) | set(third_user.keys())
    mean_value_of_movies = {key: (first_user.get(key, 0) + second_user.get(key, 0)+ third_user.get(key,0) )/ 3 for key in all_keys}
    sorted_mean_value_of_movies=sorted(mean_value_of_movies.items(), key=lambda x: x[1], reverse=True)
    return sorted_mean_value_of_movies
```

```python
sorted_predected_score_avarage_method=predected_score_avarage_method(user_recomendations_average_method)
```

# Solution for Part- (b)

The provided formula for calculating  Average Pair-wise Disagreements, involves summing the absolute differences between predicted ratings for an item given by each pair of users in the group and then multiply by the normalization 
factor:      $$\frac{2}{{|G|(|G|-1)}}$$

--------------------------------------------------------------------

The formula is:
$$
\text{dis}(G, i) = \frac{2}{{|G|(|G|-1)}} \times \sum_{u \neq v, u, v \in G} \left| \text{predictedrating}(u, i) - \text{predictedrating}(v, i) \right|
$$

---------------------------------------------------------------------
The reference of the paper from where this Pair-wise Disagreements function is taken is given below:

https://www.researchgate.net/publication/220473532_Space_efficiency_in_group_recommendation



```python
User_1 = 196
User_2 = 222
User_3 = 12

group=[User_1,User_2,User_3]
user_recomendations={}
```

```python
for user in group:
    uses_recomendation=get_recommended_movies(user, user_item_matrix, correlation_matrix)
    user_recomendations[user]=  dict(uses_recomendation)
```

```python
user_recomendations
```

```python
from itertools import combinations

def calculate_group_disagreement_for_all_items(group_predicted_ratings):
    
    users = list(group_predicted_ratings.keys())
    num_users = len(users)
    if num_users < 2:
        return {}  

 
    total_disagreements = {}


    for u, v in combinations(users, 2):
        for item in set(group_predicted_ratings[u].keys()).union(group_predicted_ratings[v].keys()):
            disagreement = abs(group_predicted_ratings[u].get(item, 0) - group_predicted_ratings[v].get(item, 0))
            total_disagreements.setdefault(item, 0)
            total_disagreements[item] += disagreement

    
    normalization_factor = 2 / (num_users * (num_users - 1))
    average_disagreements = {item: normalization_factor * disagreement for item, disagreement in total_disagreements.items()}

    average_disagreements = sorted(average_disagreements.items(), key=lambda x: x[1])
    return average_disagreements



```

```python
group_disagreements = calculate_group_disagreement_for_all_items(user_recomendations)


for i, j in group_disagreements:
    print(f"Group Disagreements for all item {i}:  {j}")
```

-------------------------------------------------------
In the upper cell we have already disagreements score for every item for all the users in the group. The more higher the disagreement score means the more disagree the group members are about the predected  movie ratings. We can use these disagreements while implimenting the Avarage method group recomendation.

When for a group , we calculate the avarage score of a movie we can subtract the  disagreement of that movie for the avarage score. In this way, we can balance the movie recomendation score for all the three users in the group.

## Movie score predection for a group of three users with Average Method and disagreements score

```python
User_1 = 196
User_2 = 222
User_3 = 12


group_average_method_with_disagreements=[User_1,User_2,User_3]
group_disagreements=dict(group_disagreements)

def predected_score_avarage_method_with_disagreement(group_average_method_with_disagreements):
    first_user= user_recomendations[User_1]
    second_user=user_recomendations[User_2]
    third_user= user_recomendations[User_3]
    all_keys=set(first_user.keys()) | set(second_user.keys()) | set(third_user.keys())
    mean_value_of_movies = {key: (first_user.get(key, 0) + second_user.get(key, 0)+ third_user.get(key,0) )/ 3 - group_disagreements.get(key,0) for key in all_keys}
    
    sorted_mean_value_of_movies=sorted(mean_value_of_movies.items(), key=lambda x: x[1], reverse=True)
    return sorted_mean_value_of_movies
```

```python
sorted_predected_score_avarage_method_with_disagreement=predected_score_avarage_method_with_disagreement(user_recomendations)
```
